```python
import pandas as pd
import rdrobust

# Load the dataset
data = pd.read_csv('/content/headstart.csv')

# Define the running variable and the outcome variable
running_var = 'povrate60'
outcome_var = 'mort_age59_all_postHS'

# Set the cutoff value for the running variable (e.g., the median poverty rate in 1960)
cutoff = data[running_var].median()

# Perform the RD analysis
rd_result = rdrobust.rdrobust(y=data[outcome_var], x=data[running_var], c=cutoff)

# Print the results
print(rd_result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


# Set the bandwidth for grouping the data
bandwidth = 7

# Group the data into categories based on the running variable
data['group'] = pd.cut(data[running_var], bins=np.arange(40, 81, bandwidth))

# Calculate the mean and standard error for each group
grouped_data = data.groupby('group')[outcome_var].agg(['mean', 'sem']).reset_index()

# Plot the raw cell means (triangles) and their 95% confidence intervals (bars)
plt.errorbar(x=grouped_data.index, y=grouped_data['mean'], yerr=1.96 * grouped_data['sem'],
             fmt='^', capsize=5, label='Raw cell means')

# Fit a quadratic model to the data (parametric estimate)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(data[[running_var]])

# Fit the model
model = LinearRegression().fit(X_poly, data[outcome_var])

# Predict values for plotting
x_range = np.linspace(data[running_var].min(), data[running_var].max(), 100)
y_pred = model.predict(poly.transform(x_range.reshape(-1, 1)))

# Plot the parametric estimate (dashed line)
plt.plot(x_range, y_pred, '--', label='Parametric estimate (quadratic)')

# Add labels and legend
plt.xlabel('1960 County Poverty Rate (%)')
plt.ylabel('High School Completion Rate')
plt.legend()

# Show the plot
plt.show()
```
